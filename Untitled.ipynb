```python
data={'Dev':[],'Acq':[],'Oth':[],'m':[],'f':[]}
aa=[54,82,51,83,91,81,128,204,195,211,238,257,274]
b=[13,14,14.2,15,14.8,19,19.3,19.6,19.2,19,18,18.2,19]
c=[.2,.25,.3,.25,.26,.27,.44,.49,.58,.48,.48,.43,.43]
d=[.1,.15,.14,.18,.15,.21,.15,.12,.12,.13,.11,.15,.17]
ff=[.311,.32,.35,.38,.38,.35,.34,.32,.27,.27,.28,.27,.27]
m=[]

count=0
for f in ff:
    m.append(round(100*(1-.2-f))/100)
print(m)

count=0
total=0
dev=0
acq=0
oth=0
for a in aa:
    #print(f'row: {count}')
    #print('--------------')
    total=a*1000*((b[count]-3)/100)
    dev=total*c[count]
    acq=total*(1-c[count]-d[count])
    oth=total-dev-acq
    
    data['Dev'].append(dev)
    data['Acq'].append(acq)
    data['Oth'].append(oth)
    data['m'].append(m[count])
    data['f'].append(ff[count])
    
    print(f'{round(dev)},{round(acq)},{round(oth)}')
    count=count+1
    
df=pd.DataFrame(data)
df
```
